```python
#import the libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Note for entering name
print("[Note: Make First letter of each word Capital]")

# Enter Actor/Actress name
actor_name = input("Enter the name of the Actor/Actress:")
formatted_name = actor_name.replace(" ", "_")

# making url as per user input
url = f"https://en.wikipedia.org/wiki/{formatted_name}_filmography"

# fetching url
response = requests.get(url)

 
if response.status_code != 200:
    url=f"https://en.wikipedia.org/wiki/{formatted_name}#filmography"  #for some names this URL works
    response = requests.get(url)
    if response.status_code != 200:
        print("The page for this actor does not exist. Please check the name and try again.")

# For extracting the info.    
soup = BeautifulSoup(response.text, 'html.parser')

titles_only = []     # List to store only movie titles

# For those who have films stored in 2nd list in Filmography section and have films title in <th>
if not titles_only:
    film_table = soup.find_all('table', class_='wikitable')[1]
    
    rows = film_table.find_all('tr')[1:]  # Skip the header row
    current_year = None  # To keep track of the current year
    
    for row in rows:    
        title_cell = row.find('th')  # Get the <th> element for the title
        if title_cell:
            film_name = title_cell.get_text(strip=True)  # Extract the title
            titles_only.append(film_name)  # Only the title
            

# ---------------------------------------------------------------
# For those who have films stored in 1st list in Filmography section and have films title in <th>

if not titles_only:
    film_table = soup.find_all('table', class_='wikitable')[0]
    
    rows = film_table.find_all('tr')[1:]  # Skip the header row
    for row in rows:
        cols = row.find_all('td')  # Get the <td> elements for titles
        if cols:
            film_name = cols[0].get_text(strip=True)  # Extract the title from the first <td>
            titles_only.append(film_name)
# ---------------------------------------------------------------
# For those who have films titles in <td>


if not titles_only:
    film_table = soup.find_all('table', class_='wikitable')[1]
    
    rows = film_table.find_all('tr')[1:]  # Skip the header row
    for row in rows:
        cols = row.find_all('td')  # Get the <td> elements for titles
        if cols:
            film_name = cols[0].get_text(strip=True) 
            titles_only.append(film_name)
                

years = [str(year) for year in range(1900, 2025)]  # Convert each year to a string
j = ', '.join(years)
for i in titles_only:
    if i in j:
        film_table = soup.find_all('table', class_='wikitable')[1]


        titles_only = []
        
        rows = film_table.find_all('tr')[1:]  # Skip the header row
        for row in rows:
            cols = row.find_all('td')  # Get the <td> elements for titles
            if cols:
                film_name = cols[0].get_text(strip=True)
                titles_only.append(film_name)
                
                
titles_only.reverse()  #Chronological descending order
    
print(f"Films done by {actor_name} in descending order:")
print("(†) This indicated To be released")
for film in titles_only:
    print(f"- {film}")
```
